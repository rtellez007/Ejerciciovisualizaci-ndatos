```python
# Carga Fuentes de Datos
import pandas as pd
filename = '/content/drive/MyDrive/CodingDojo/Data/RailsToTrails_National Count Data_week.xlsx'
df = pd.read_excel(filename)
df
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
```

```python

dfAtri = df.loc[(df['Year'] == 2019),:]
Semanas = list(dfAtri['Timeframe'])
Bici2019 =  list(dfAtri['Bikes14counters'])

dfAtri = df.loc[(df['Year'] == 2020),:]
Bici2020=  list(dfAtri['Bikes14counters'])
```

```python
Semanas
```

```python

fig, ax = plt.subplots()
ax.plot(Semanas, Bici2019, color="orange", label="2019")
ax.plot(Semanas, Bici2020, color="purple", label="2020")
ax.title.set_text('Venta de Unidades de Bicletas en Pandemia')
ax.set_xlabel('Semanas', color = 'b',family='serif', weight='normal', size = 10, labelpad = 10) 
ax.set_ylabel('Cantidad Unidades Vendidas', color = 'g', family='serif', weight='normal', size = 10, labelpad = 10) 
plt.xticks(rotation=90,fontsize=6)
plt.yticks(fontsize=8)
leg = ax.legend()

plt.show()
```

```python
dfAtri = df.loc[(df['Year'] == 2019),:]
Semanas = list(dfAtri['Timeframe'])
Bici2019 =  list(dfAtri['Bikes14counters'].cumsum())

dfAtri = df.loc[(df['Year'] == 2020),:]
Bici2020=  list(dfAtri['Bikes14counters'].cumsum())
Bici2019
```

```python

fig, ax = plt.subplots()
ax.bar(Semanas, Bici2019, color="blue", label="2019")
ax.plot(Semanas, Bici2020, color="red", label="2020")
ax.title.set_text('Acumulado de Unidades de Bicletas en Pandemia')
ax.set_xlabel('Semanas', color = 'b',family='serif', weight='normal', size = 10, labelpad = 10) 
ax.set_ylabel('Cantidad Unidades Vendidas', color = 'g', family='serif', weight='normal', size = 10, labelpad = 10) 
plt.xticks(rotation=90,fontsize=6)
plt.yticks(fontsize=8)
leg = ax.legend()

plt.show()
```
